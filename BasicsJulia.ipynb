```julia
using CSV
using DataFrames
using GLM

# Read the Iris dataset into a DataFrame
iris = CSV.read("Iris.csv", DataFrame, normalizenames=true)

# Assuming you want to predict 'SepalLengthCm' based on 'PetalLengthCm'
# Assign the predictor and response variables
X = iris.PetalLengthCm
y = iris.SepalLengthCm

# Fit a linear regression model
model = lm(@formula(SepalLengthCm ~ PetalLengthCm), iris)

# Print the summary of the linear regression model
println(summary(model))

# Get the coefficients of the model
coef_values = coef(model)
println("Coefficients: $coef_values")

new_data = DataFrame(PetalLengthCm = [3.5, 4.0, 4.5])  # Replace these values with the ones you want to predict
predicted_values = predict(model, new_data)

# Display predicted values
println("Predicted SepalLengthCm values:")
println(predicted_values)
```

```julia
using Random
function train_test_split(X, y, train_size)
    n = nrow(X)
    indices = shuffle(1:n)  # Shuffle indices instead of using randperm
    train_indices = indices[1:round(Int, train_size * n)]
    test_indices = setdiff(indices, train_indices)
    return (X[train_indices, :], y[train_indices], X[test_indices, :], y[test_indices])
end
```

```julia
X_train, y_train, X_test, y_test = train_test_split(features, target, 0.8)
```

```julia
using DataFrames, CSV, DecisionTree, Random

# Load the dataset (if not already loaded)
# Replace 'Iris.csv' with the correct path to your dataset
iris = CSV.read("Iris.csv", DataFrame)

# Define the features and target
features = Matrix(iris[:, 1:4]) # Assuming the features are in columns 1 to 4
target = iris[:, 5] # Assuming the target column is the 5th column

# Split the data into training and testing sets
Random.seed!(123) # Set seed for reproducibility
train_rows = randsubseq(1:size(features, 1), 0.7) # 70% for training
X_train, y_train = features[train_rows, :], target[train_rows]
X_test, y_test = features[setdiff(1:end, train_rows), :], target[setdiff(1:end, train_rows)]

# Train a random forest classifier
model = DecisionTree.RandomForestClassifier(n_trees=10, max_depth=5)
DecisionTree.fit!(model, X_train, y_train)

# Make predictions on the test set
predictions = DecisionTree.predict(model, X_test)

# Calculate accuracy
accuracy = sum(predictions .== y_test) / length(y_test)
println("Accuracy: $accuracy")
```

```julia
using DataFrames, Random, DecisionTree

# Creating a DataFrame manually for demonstration purposes
# You should load your actual dataset instead
data = DataFrame(
    Id = 1:150,
    SepalLengthCm = rand(4.3:0.1:7.9, 150),
    SepalWidthCm = rand(2.0:0.1:4.4, 150),
    PetalLengthCm = rand(1.0:0.1:6.9, 150),
    PetalWidthCm = rand(0.1:0.1:2.5, 150),
    Species = repeat(["Iris-setosa", "Iris-versicolor", "Iris-virginica"], inner=50)
)

# Define features and target
features = data[:, Not(:Id, :Species)]
target = data[:, :Species]

# Split the data into training and testing sets
Random.seed!(123) # Set seed for reproducibility
train_indices = randsubseq(1:size(data, 1), 0.8)
test_indices = setdiff(1:size(data, 1), train_indices)

X_train = features[train_indices, :]
y_train = target[train_indices]

X_test = features[test_indices, :]
y_test = target[test_indices]

# Train a random forest classifier
model = DecisionTree.RandomForestClassifier(n_trees=10, max_depth=5)
DecisionTree.fit!(model, Matrix(X_train), y_train)

# Make predictions on the test set
predictions = DecisionTree.predict(model, Matrix(X_test))

# Calculate accuracy
accuracy = sum(predictions .== y_test) / length(y_test)
println("Accuracy: $accuracy")
```
